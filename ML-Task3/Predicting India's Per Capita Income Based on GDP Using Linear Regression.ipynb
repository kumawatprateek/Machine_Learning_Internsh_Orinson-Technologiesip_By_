# **Predicting India's Per Capita Income Based on GDP Using Linear Regression**

**Step 1: Import necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

**Seep 2: Load Dataset**

```python
df = pd.read_csv('India_GDP_Data.csv')
df
```

**Step 3: Split data into features (X) and target (y)**

```python
X=df[['GDP_In_Billion_USD']]# Feature
Y=df[['Per_Capita_in_USD']]# Target
```

**Step 4: Split the dataset into training and testing sets (80% train, 20% test)**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

**Step 5: Create and train the Linear Regression model**

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

**Step 6: Make predictions**

```python
y_pred = model.predict(X_test)
model.score(X_test,y_test)
```

**Step 7: Evaluate the model performance**

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```

**Step 8: Visualize the regression line with the data points**

```python
plt.scatter(X, Y, color='blue')  # Plot data points
plt.plot(X, model.predict(X), color='red', linewidth=2)  # Plot regression line
plt.title('Pridict Per capita')
plt.xlabel('GDP_In_Billion_USD')
plt.ylabel('Per_Capita_in_USD')
plt.show()
```

**Step 9: Predict Per Capita for a specific GDP value (e.g., 10,000 Billion USD)**

```python
n=int(input("Enter the GDP value (e.g., 10,000 Billion USD): "))
m=model.predict([[n]])
print("The Per Capita (USD): ",m[0])
```
